```python
#Venue Code = 0 means away and 1 means home
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
#replaces colon and minutes from time with nothing
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"]=matches["date"].dt.dayofweek
```

```python
#True is 1 and False is 0
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
#doesn't take in higher value opp_codes as more difficult opponents
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=30, random_state=1)
```

```python
train = matches[matches["date"] < '2024-01-01']
```

```python
test = matches[matches["date"] > '2024-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code", "sh", "sot", "poss", "xg", "xga"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
predictions = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test["target"], predictions)
```

```python
#Rolling Averages to imporve precision of the model
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester United").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(4, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt", "season"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors], train["target"])
    predictions = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=predictions), index=test.index)
    precision = precision_score(test["target"], predictions)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
